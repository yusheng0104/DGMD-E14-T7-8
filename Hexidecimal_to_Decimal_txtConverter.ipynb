```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('test.txt', sep=" ", skiprows=1, header=None, dtype = str)
```

```python
data
```

```python
output_df = pd.DataFrame()
output_df["Hanldes"] = data[0] +" "+ data[1] +" " +data[2] +" "+ data[3] + data[4]
output_df["Timestamps (sec)"] = data[6] + data[5]
output_df["Pressure (mbar)"] = data[10] + data[9] + data[8] + data[7]
output_df["Temperature (degree C)"] = data[12] + data[11]
```

```python
output_df["Timestamps (sec)"]  = [int(x, 16)/100 for x in  output_df["Timestamps (sec)"]]
output_df["Pressure (mbar)"]  = [int(x, 16)/100 for x in  output_df["Pressure (mbar)"]]
output_df["Temperature (degree C)"]  = [int(x, 16)/10 for x in  output_df["Temperature (degree C)"]]
```

```python
output_df.to_csv("test.csv", index = False)
```
